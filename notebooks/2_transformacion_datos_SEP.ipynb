```python
import pandas as pd
import glob
import os
```

```python
csv_files = sorted(glob.glob("../data/raw/formato_911/formato_911_basica_*.csv"))
csv_files
```

```python
dataframes = {}

for file in csv_files:
    # Extraer el nombre del archivo sin extensión
    year = os.path.splitext(os.path.basename(file))[0].replace("formato_911_basica_", "")
    
    # Leer el CSV
    df = pd.read_csv(file)
    
    # Guardar en el diccionario
    dataframes[year] = df

# Ejemplo de acceso
dataframes["2022-2023"].head()
```

```python
for year, df in dataframes.items():
    print(year, df.shape)
```

Todos los dataframes tienen 156 columnas, aunque difieren en el número de filas.

```python
# Combinar todos los DataFrames y añadir la columna 'periodo_escolar'
combined_df = pd.concat(
    [df.assign(periodo_escolar=year) for year, df in dataframes.items()],
    ignore_index=True
)
```

```python
combined_df.info()
combined_df.head()
```

```python
#Verificando los valores de la columna "periodo_escolar"
combined_df['periodo_escolar'].unique()
```

Ahora los 6 archivos .csv están cargados todos en un solo DataFrame, combined_df, y existe una nueva columna identificadora llamada "periodo_escolar".

```python
import os

# Subir un nivel (del notebook a la raíz del proyecto)
output_dir = os.path.join("..", "data", "processed")
os.makedirs(output_dir, exist_ok=True)
```

```python
# Guardar el DataFrame combinado en CSV dentro de ../data/processed
output_path = os.path.join(output_dir, "sep_datos_tidy.csv")
combined_df.to_csv(output_path, index=False, encoding="utf-8-sig")

print(f"✅ Archivo guardado en: {output_path}")
print(f"Total de filas: {len(combined_df)}")
```
